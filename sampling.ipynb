```python
import pandas as pd
import random
```

```python
### Sampling: taking 100000 lines from train.txt
trainfile = 'train.txt'

n = 2847802
s = 500000
skip = sorted(random.sample(xrange(n),n-s))

train = pd.read_csv(trainfile, header = None, sep = '\t', skiprows = skip)

train.head()
```

```python
### Write sampled lines to txt file
train.to_csv('train_sample.txt', sep = '\t', header = None, index = None)
```
